# 4. 문서 청크로 나누기

청킹을 완수하면 청킹된 내용을 상위 10개까지 출력하고, 각 청킹방식과 parameter의 뜻을 markdow으로 정리 했습니다. 

## i. CharacterTextSplitter 

```python
from langchain.text_splitter import CharacterTextSplitter

text_splitter = CharacterTextSplitter(
    separator="\n\n",
    chunk_size=100,
    chunk_overlap=10,
    length_function=len,
    is_separator_regex=False,
)

splits = text_splitter.split_documents(docs)
```

```python
# 청킹된 내용의 상위 10개 출력
print("상위 10개 청킹 결과:")
for i, chunk in enumerate(splits[:10], 1):
    print(f"청크 {i}: {chunk.page_content}")
```

## ii. RecursiveCharacterTextSplitter

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter

recursive_text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=100,
    chunk_overlap=10,
    length_function=len,
    is_separator_regex=False,
)

splits = recursive_text_splitter.split_documents(docs)
```

```python
# 청킹된 내용의 상위 10개 출력
print("상위 10개 청킹 결과:")
for i, chunk in enumerate(splits[:10], 1):
    print(f"청크 {i}: {chunk.page_content}")
```

## 1. `CharacterTextSplitter`
- **기본적인 문자 기반 분할기**입니다.
- 지정한 `separator` 문자열을 기준으로 텍스트를 분할합니다. 
  - 예: `separator="\n\n"`으로 지정하면 두 줄 띄어쓰기를 기준으로 텍스트를 나눕니다.
- 분할 과정에서 `chunk_size`와 `chunk_overlap`을 기준으로 최대 청크 크기와 청크 간 중첩 범위를 설정할 수 있습니다.
- 주어진 `separator`가 텍스트 내에 존재하지 않는 경우, 텍스트 전체를 단일 청크로 처리할 수 있습니다.

### 장점
- **간단하고 빠름**: 단일 구분자(`separator`)를 사용해 청킹하므로, 분할 과정이 간단하고 빠릅니다.
- **예측 가능한 분할**: 특정 구분자를 기반으로 나누기 때문에, 분할 방식이 예측 가능하고 단순한 경우에 적합합니다.
- **설정이 쉬움**: 구분자와 청크 크기만 설정하면 되므로, 간단한 분할이 필요한 경우 쉽게 사용할 수 있습니다.

### 단점
- **유연성 부족**: 하나의 구분자만 사용할 수 있어 다양한 텍스트 형식이나 복잡한 텍스트 구조에 적합하지 않을 수 있습니다.
- **청크 크기 일관성 부족**: 지정한 구분자가 없거나 적절한 위치에 없을 경우, 청크 크기가 일관되지 않게 분할될 수 있습니다.
- **복잡한 텍스트에 비효율적**: 여러 분할 기준이 필요한 복잡한 문서에서는 분할의 효과가 떨어질

## CharacterTextSplitter 파라미터 설명
### `separator`
- **설명**: 텍스트를 분할할 구분자를 설정합니다. 예를 들어, `"\n\n"`으로 설정하면 두 줄의 띄어쓰기를 기준으로 텍스트가 분할됩니다.
- **기본값**: `None`
- **사용 예시**: 
  - `separator="\n\n"`: 두 줄 띄어쓰기를 기준으로 텍스트를 분할.

### `chunk_size`
- **설명**: 각 청크의 최대 크기를 설정합니다. 텍스트를 지정된 `chunk_size` 이하로 나누어 최대 크기를 지정합니다.
- **기본값**: 설정하지 않으면 기본값이 지정될 수 있습니다.
- **사용 예시**: 
  - `chunk_size=100`: 각 청크의 크기가 100글자를 넘지 않도록 분할됩니다.

### `chunk_overlap`
- **설명**: 청크 간에 겹치는 부분의 크기를 설정합니다. `chunk_overlap` 값을 설정하면, 각 청크의 끝과 다음 청크의 시작 부분이 겹치는 양을 지정할 수 있습니다.
- **기본값**: 0
- **사용 예시**:
  - `chunk_overlap=10`: 두 청크가 10글자씩 겹쳐서 분할됩니다.

### `length_function`
- **설명**: 텍스트 길이를 측정하는 함수입니다. 기본적으로 `len()` 함수를 사용하지만, 텍스트의 길이를 측정하는 다른 함수를 지정할 수 있습니다.
- **기본값**: `len`
- **사용 예시**: 
  - `length_function=len`: 기본적으로 문자 수로 길이를 계산합니다.

### `is_separator_regex`
- **설명**: 구분자(`separator`)가 **정규식(Regex)** 인지 여부를 설정하는 파라미터입니다. `True`로 설정하면 구분자가 정규식 패턴으로 해석됩니다.
- **기본값**: `False`
- **사용 예시**:
  - `is_separator_regex=True`: 구분자를 정규식으로 해석하고, 여러 패턴을 사용할 수 있습니다.

### 장점
- **유연한 분할**: 여러 구분자를 순차적으로 적용하여, 텍스트를 재귀적으로 나누므로 복잡한 구조에서도 최적의 분할을 할 수 있습니다.
- **청크 크기 일관성**: 다양한 구분자를 통해 최대한 균일하게 청크 크기를 유지할 수 있어, 텍스트를 일정한 크기로 다루기에 좋습니다.
- **복잡한 텍스트에 적합**: 문단, 문장, 단어 단위로 재귀적 분할이 가능해 복잡한 텍스트 구조에서도 효율적으로 청킹할 수 있습니다.

- **설정 복잡성**: 여러 구분자를 설정해야 하므로 간단한 경우보다 설정이 복잡해질 수 있습니다.

## `RecursiveCharacterTextSplitter` 파라미터 설명

### `chunk_size`
- **설명**: 각 청크의 최대 크기를 설정합니다. 텍스트를 `chunk_size`만큼 나누어 최대 크기를 지정합니다.
- **기본값**: 설정하지 않으면 기본값이 지정될 수 있습니다.
- **사용 예시**: 
  - `chunk_size=100`: 각 청크의 크기가 100글자를 넘지 않도록 분할됩니다.

### `chunk_overlap`
- **설명**: 청크 간에 겹치는 부분의 크기를 설정합니다. `chunk_overlap` 값을 설정하면, 각 청크의 끝과 다음 청크의 시작 부분이 겹치는 양을 지정할 수 있습니다.
- **기본값**: 설정하지 않으면 기본값이 0이 될 수 있습니다.
- **사용 예시**: 
  - `chunk_overlap=10`: 두 청크가 10글자씩 겹치게 하여 연속성을 유지합니다.

### `length_function`
- **설명**: 각 텍스트 조각의 길이를 측정하는 데 사용할 함수입니다. 기본적으로 `len()` 함수가 사용됩니다.
- **기본값**: `len`
- **사용 예시**: 
  - `length_function=len`: 기본적으로 문자 수로 길이를 계산합니다.

### `is_separator_regex`
- **설명**: 구분자(`separator`)가 **정규식(Regex)** 인지 여부를 설정하는 파라미터입니다. 
  - `True`로 설정하면 구분자가 정규식 패턴으로 해석됩니다.
  - `False`로 설정하면 구분자가 일반 문자열로 취급됩니다.
- **기본값**: `False`
- **사용 예시**: 
  - `is_separator_regex=False`: 구분자를 일반 문자열로 취급합니다.

### 추가적인 파라미터(주로 `RecursiveCharacterTextSplitter`에서 사용)

### `separators`
- **설명**: 텍스트 분할 시 사용할 여러 구분자들을 리스트 형식으로 지정할 수 있습니다. 이 파라미터는 `separator`와는 달리, 여러 개의 구분자들을 처리할 수 있도록 합니다.
- **기본값**: `['\n', ' ', ',']` 등으로 설정할 수 있습니다.
- **사용 예시**: 
  - `separators=["\n\n", ".", ","]`: 문단 구분, 마침표, 쉼표를 기준으로 텍스트를 분할합니다.

### `max_split_length`
- **설명**: 각 분할 기준에 대해 최대 분할 길이를 설정할 수 있습니다.
- **기본값**: 설정하지 않으면 기본값이 사용됩니다.
- **사용 예시**: 
- `max_split_length=200`: 각 텍스트 조각이 최대 200글자가 되도록 분할합니다.

## 9. RAG 체인 구성
LangChain의 모델과 프롬프트를 연결하여 RAG 체인을 구성합니다. 

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough

# 프롬프트 템플릿 정의
contextual_prompt = ChatPromptTemplate.from_messages([
    ("system", "Answer the question using only the following context."),
    ("user", "Context: {context}\\n\\nQuestion: {question}")
])

class DebugPassThrough(RunnablePassthrough):
    def invoke(self, *args, **kwargs):
        output = super().invoke(*args, **kwargs)
        print("Debug Output:", output)
        return output
# 문서 리스트를 텍스트로 변환하는 단계 추가
class ContextToText(RunnablePassthrough):
    def invoke(self, inputs, config=None, **kwargs):  # config 인수 추가
        # context의 각 문서를 문자열로 결합
        context_text = "\n".join([doc.page_content for doc in inputs["context"]])
        return {"context": context_text, "question": inputs["question"]}

# RAG 체인에서 각 단계마다 DebugPassThrough 추가
rag_chain_debug = {
    "context": retriever,                    # 컨텍스트를 가져오는 retriever
    "question": DebugPassThrough()        # 사용자 질문이 그대로 전달되는지 확인하는 passthrough
}  | DebugPassThrough() | ContextToText()|   contextual_prompt | model
```

## 10. 챗봇 구동 확인
질문에 응답하는 챗봇을 구동하여 질문합니다.
같은 질문을 일반 chat gpt 혹은 Gemini에 질문해보고 답변을 비교해봤고, 왜 RAG이 필요한지 간단히 markdown으로 서술해보았습니다.

```python
while True:
    print("========================")
    query = input("질문을 입력하세요 (종료하려면 'exit' 또는 'quit' 입력): ")
    
    if query.lower() in ['exit', 'quit']:
        print("프로그램을 종료합니다.")
        break
    
    response = rag_chain_debug.invoke(query)
    print("Final Response:")
    print(response.content)
```

### **RAG기법으로 생성한 챗봇의 답변** :

### **RAG(Recovery-Augmented Generation) 필요성**
RAG 기법은 기존의 대규모 언어 모델이 제공하는 정보에 기반하여 정확성을 높이고, 추가적인 배경 지식 및 구체적인 예시를 더해주는 방식입니다. 위 답변에서 RAG 모델은 추가적인 상세 정보와 구체적인 예시를 제공하여, **"법인의 주요 주주나 임원"**과 같은 내용을 포함하고 있습니다. 이는 더 실용적인 정보를 제공하는데 유용하며, 특히 법적 혹은 세무적 문제에서는 더 정확한 해석을 요구할 수 있습니다.

반면, 기본적으로 생성된 모델(ChatGPT)은 법적 언어가 일부 생략되거나 구체성이 떨어질 수 있어, 실제 세무 상황에서 해석이 모호할 수 있습니다.

따라서 RAG 기법은 정확성과 구체성을 추가하여 챗봇의 답변이 더 유용하고 실용적으로 개선될 수 있게 도와줍니다.
